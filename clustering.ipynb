```python
# Импортируем библиотеки
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd

# Создаем датафрейм
df = pd.read_csv("housing.csv")
samples = df.drop(columns=['MedInc', 'AveRooms','AveBedrms','AveOccup','Latitude','Longitude','MedHouseVal']).values

kmeans = KMeans(n_clusters = 5 , init = "k-means++" , random_state = 10)
y_kmeans = kmeans.fit_predict(samples)


plt.scatter(samples[y_kmeans == 0, 0], samples[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(samples[y_kmeans == 1, 0], samples[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(samples[y_kmeans == 2, 0], samples[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(samples[y_kmeans == 3, 0], samples[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(samples[y_kmeans == 4, 0], samples[y_kmeans == 4, 1], s = 100, c = 'purple', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters')
plt.xlabel('House age')
plt.ylabel('Population')
plt.legend()
plt.show()
```
